```python
forecast = model_fit.forecast(steps=10)
print(forecast)
```

```python
last_value = timeseries.iloc[-1]

forecast_cumsum = forecast.cumsum()
print(forecast_cumsum)
```

SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarima_model = SARIMAX(timeseries, order=(2,1,2), seasonal_order=(0,0,0,0))
sarima_model_fit = sarima_model.fit()

print("SARIMAX Model Summary:")
print(sarima_model_fit.summary())
```

```python
sarima_forecast_diff = sarima_model_fit.forecast(steps=10)
```

```python
last_index = india['Total Population, as of 1 July (thousands)'].iloc[-1]
forecast_value = last_index + sarima_forecast_diff.cumsum()

sarima_forecast_index = pd.RangeIndex(start=india.index[-1] + 1, stop=india.index[-1] + 1 + 10)
sarima_forecast_series = pd.Series(forecast_value.values, index=sarima_forecast_index)

print("\nSARIMA Forecast (Original Scale):")
print(sarima_forecast_series)

```

```python

plt.figure(figsize=(12, 6))
plt.plot(india['Total Population, as of 1 July (thousands)'], label='Original Population')
plt.plot(sarima_forecast_series, label='SARIMA Forecast', color='red')
plt.title('Total Population Forecast using SARIMA Model')
plt.xlabel('Year')
plt.ylabel('Population (thousands)')
plt.legend()
plt.grid(True)
plt.show()
```
